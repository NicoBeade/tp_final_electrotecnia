```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

#(start, end, n)
x = np.linspace(0, 2, 1000)
y1 = np.sin(2*np.pi*x)               #y = sin(2pix)
y2 = np.cos(2*np.pi*x)

plt.plot(x,y1, 'm.', label = '$sin(2\pi x)$')
plt.plot(x,y2, 'r', label = '$cos(2\pi x)$')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 2, 1000)
y1 = np.sin(2*np.pi*x)               #y = sin(2pix)
y2 = np.cos(2*np.pi*x)

plt.figure(figsize=(12, 4))
plt.subplot(2,1,1)          #2filas, 1 columna, posicion 1
plt.plot(x,y1, 'b-')

plt.subplot(2,1,2)          #2filas, 1 columna, posicion 2
plt.plot(x,y2, 'r-')

plt.tight_layout()
plt.show()

#Al reves

plt.subplot(1,2,1)          #1 fila, 2 columnas, posicion 1
plt.plot(x,y1, 'b-')

plt.subplot(1,2,2)          #1 fila, 2 columna, posicion 2
plt.plot(x,y2, 'r-')

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 2, 1000)
y1 = np.sin(2*np.pi*x)               #y = sin(2pix)
y2 = np.cos(2*np.pi*x)

plt.figure(figsize=(12, 4))

plt.subplot(2,1,1)          #2filas, 1 columna, posicion 1
plt.title("Grafico azul")
plt.xlabel("$Tiempo [\mu seg]$")
plt.ylabel("$Tension [V]$")
plt.plot(x,y1, 'b-')

plt.subplot(2,1,2)          #2filas, 1 columna, posicion 2
plt.title("Grafico rojo")
plt.xlabel("$Tiempo [\mu seg]$")
plt.ylabel("$Tension [V]$")
plt.plot(x,y2, 'r-')

plt.axis([0, 3, -1, 1])     #[xmin, xmax, ymin, ymax]
plt.grid()

plt.tight_layout()
plt.show()
```

```python
#Histograma

mu = 0
sigma = 0.5
var = sigma**2
n = 1000
bins = 20

t = np.linspace(-10, 10, 1000)
dist_normal = 1/np.sqrt(2*np.pi)*sigma*np.exp(-(t-mu)**2/(2*sigma**2))


rand = np.random.normal(mu, var, n) #media 0, desvio estandar 2, 1000 puntos
plt.hist(rand, bins, density = True)  #20 bins
plt.plot(t, dist_normal)
plt.show()
```
